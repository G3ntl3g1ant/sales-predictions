Searching through any data frames, the first thing that must be identified is if there is any missing data, to begin:

```python
sp2 = pd.read_csv(filename)
sp2.set_index('Item_Identifier', inplace=True)
sp2.head(2)
```

4) Now to see if there are any missing values.

#### 5) First there is missing data from the **'Item_Weight'** column, because there are so many entries missing and assuming the weight of these items could possibly taint the results, for now **'Missing'** will be the replacement variable until there is data to replace the missing information, or if it is found to be unnecessary. 

```python
# What values need to be changed
sp2['Item_Weight'].fillna('Missing', inplace = True)
# Checking to see what other items are missing
sp2.isna().sum()
```

Now all that is left is the **'Outlet_Size'** column to fix, first the values for **'Outlet_Size'** must be identified

```python
# Checking the values of the 'Outlet_Size' column
sp2['Outlet_Size'].value_counts()
```

Now that the values are stated we can plug in the average outlet size to fill in the rest of the missing data. 

First, the values are changed to an int.

```python
# Values to be changed
replace_values = {'Small' : 1, 'Medium' :2, 'High': 3}
sp2['Outlet_Size'] = sp2['Outlet_Size'].replace(replace_values)
sp2['Outlet_Size'].value_counts()
```

Then, to change the missing data into the **median** for **'Outlet_Size'**

```python
# Now to look for the median of the oulet sizes
# mos = median outlet size
mos = sp2['Outlet_Size'].median()
sp2['Outlet_Size'].fillna(mos, inplace=True)
```

And finally, change into an int.

```python
sp2['Outlet_Size'] = sp2['Outlet_Size'].astype(int)
sp2.info()
```

#### 6) Now to check the data frame once again for any missing values.

```python
sp2.isna().sum()
```

#### Now with that out of the way, to check for needed columns and what data might be unnecessary for the data frame/data collection

```python
sp2.nunique()
```

After looking at the data as well as the key to the data, the **'Outlet_Establishment_Year'** column is unnecessary to the information we are looking for. So, it shall be removed.

```python
sp2.drop(columns = {'Outlet_Establishment_Year'}, inplace=True)
sp2.head(2)
```

To finish the last bit of cleaning, it seems odd to have 5 different values for the **'Item_Fat_Content'** column when generally it should have only 2 values, but to find the truth, the values need to be checked.

```python
sp2['Item_Fat_Content'].value_counts()
```

So indeed it has different values stating the same thing, so the values will be replaced with the correct versions.

```python
# LF changed into Low Fat
sp2['Item_Fat_Content'] = sp2['Item_Fat_Content'].replace('LF', 'Low Fat')
# reg changed into Regular
sp2['Item_Fat_Content'] = sp2['Item_Fat_Content'].replace('reg', 'Regular')
# low fat changed into Low Fat
sp2['Item_Fat_Content'] = sp2['Item_Fat_Content'].replace('low fat', 'Low Fat')
# Checking that all values were changed correctly
sp2['Item_Fat_Content'].value_counts()
```

```python
# Checking the Item_Fat_Content column
sp['Item_Fat_Content'].value_counts()
```

Lets go ahead and fix these repeats/errors and organize the **ordinal** data since it is an easy fix.

```python
# Ordinal Encoding 'Item_Fat_Content' 
# replacement dictionary = rep_dic
rep_dic = {'Low Fat':0, 'LF':0, 'low fat': 0, 'Regular':1, 'reg':1}
sp['Item_Fat_Content'].replace(rep_dic, inplace=True)
sp['Item_Fat_Content'].value_counts()
```

```python
# Checking the values of Outlet_Size
sp['Outlet_Size'].value_counts()
```

```python
# Ordinal Encoding 'Outlet_Size' 
# replacement dictionary = rep_dic
rep_dic = {'Medium':1, 'Small':0, 'High':2}
sp['Outlet_Size'].replace(rep_dic, inplace=True)
sp['Outlet_Size'].value_counts()
```

```python
sp['Outlet_Location_Type'].value_counts()
```

```python
# Ordinal Encoding 'Outlet_Location_Type' 
# replacement dictionary = rep_dic
rep_dic = {'Tier 3':2, 'Tier 2': 1, 'Tier 1':0}
sp['Outlet_Location_Type'].replace(rep_dic, inplace=True)
sp['Outlet_Location_Type'].value_counts()
```

```python
# Ordinal Encoding 'Outlet_Type' 
# replacement dictionary = rep_dic
rep_dic = {'Supermarket Type1':1, 'Grocery Store': 0, 'Supermarket Type3':3, 'Supermarket Type2':2}
sp['Outlet_Type'].replace(rep_dic, inplace=True)
sp['Outlet_Type'].value_counts()
```

Now with that done, it is time to identify our **X** and **y**!

## Defining X and y

```python
sp.info()
```

For our *target* we are going to focus in on our **Item_Outlet_Sales** column, while our features will be **every other column**. 

With what we are looking for it is unnecessary for us to keep the ** Item_Identifier**, **Outlet_Identifier** and the **Outlet_Establishment_Year** columns, due to that they are too unique in their values and do not add any additional valuable information for our specific target. 

```python
sp = sp.drop(['Item_Identifier','Outlet_Identifier', 'Outlet_Establishment_Year'], axis=1)
sp.head()
```

```python
# X and y values
X = sp.drop('Item_Outlet_Sales', axis=1)
y = sp['Item_Outlet_Sales']
```

## Train_Test_Split

```python
# Performing a train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Making a Preprocessing Pipelines

```python
# Creating categorical and numeric selectors
# categorical selector = cat_sel
cat_sel = make_column_selector(dtype_include='object')
# numerical selector = num_sel
num_sel = make_column_selector(dtype_include='number')
```

For our provided data, we are going to use the **mean** imputer for the **numerical missing data** and the **most_frequent** imputer for the **nominal missing data**. 

As well we shall add in our StandardScaler() and OneHotEncoder() methods here.

```python
# Imputers
freq_imp = SimpleImputer(strategy='most_frequent')
mean_imp = SimpleImputer(strategy='mean')
# Scaler
scaler = StandardScaler()
# OneHotEncoder 
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)
```

```python
# Numeric pipeline
num_pipe = make_pipeline(mean_imp, scaler)
num_pipe
```

```python
# Categorical Pipeline
cat_pipe = make_pipeline(freq_imp, ohe)
cat_pipe
```

## ColumnTransformer

```python
# Tuples for Column Transformer
num_tup = (num_pipe, num_sel)
cat_tup = (cat_pipe, cat_sel)
# ColumnTransformer
preprocessor = make_column_transformer(num_tup, cat_tup)
preprocessor
```

## Transformer Data

Now to finalize everything by processing the data and then of course, checking our work.

```python
# fit on training data
preprocessor.fit(X_train)
# transform train and test data
X_train_processed = preprocessor.transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
# Checking for missing values
print(np.isnan(X_train_processed).sum().sum())
print(np.isnan(X_test_processed).sum().sum())
```
