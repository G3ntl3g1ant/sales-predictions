First there is missing data from the **'Item_Weight'** column, because there are so many entries missing and assuming the weight of these items could possibly taint the results, for now **'Missing'** will be the replacement variable until there is data to replace the missing information, or if it is found to be unnecessary.

And finally, change into an int.

```python
sp2['Outlet_Size'] = sp2['Outlet_Size'].astype(int)
sp2.info()
```

Now to check the data frame once again for any missing values.

```python
sp2.isna().sum()
```

### Ordinal Encoding

```python
# Checking the Item_Fat_Content column
sp['Item_Fat_Content'].value_counts()
```

Lets go ahead and fix these repeats/errors and organize the **ordinal** data since it is an easy fix.

```python
# Ordinal Encoding 'Item_Fat_Content' 
# replacement dictionary = rep_dic
rep_dic = {'Low Fat':0, 'LF':0, 'low fat': 0, 'Regular':1, 'reg':1}
sp['Item_Fat_Content'].replace(rep_dic, inplace=True)
sp['Item_Fat_Content'].value_counts()
```

Perfect! Now I will go ahead and take the other **ordinal** data available and make a similar fix. 

```python
# Checking the values of Outlet_Size
sp['Outlet_Size'].value_counts()
```

```python
# Ordinal Encoding 'Outlet_Size' 
# replacement dictionary = rep_dic
rep_dic = {'Medium':1, 'Small':0, 'High':2}
sp['Outlet_Size'].replace(rep_dic, inplace=True)
sp['Outlet_Size'].value_counts()
```

```python
sp['Outlet_Location_Type'].value_counts()
```

```python
# Ordinal Encoding 'Outlet_Location_Type' 
# replacement dictionary = rep_dic
rep_dic = {'Tier 3':2, 'Tier 2': 1, 'Tier 1':0}
sp['Outlet_Location_Type'].replace(rep_dic, inplace=True)
sp['Outlet_Location_Type'].value_counts()
```

```python
sp['Outlet_Type'].value_counts()
```

```python
# Ordinal Encoding 'Outlet_Type' 
# replacement dictionary = rep_dic
rep_dic = {'Supermarket Type1':1, 'Grocery Store': 0, 'Supermarket Type3':3, 'Supermarket Type2':2}
sp['Outlet_Type'].replace(rep_dic, inplace=True)
sp['Outlet_Type'].value_counts()
```

Now with that done, it is time to identify our **X** and **y**!

## Defining X and y

```python
sp.info()
```

For our *target* we are going to focus in on our **Item_Outlet_Sales** column, while our features will be **every other column**. 

With what we are looking for it is unnecessary for us to keep the **Item_Identifier**, **Outlet_Identifier** and the **Outlet_Establishment_Year** columns, due to that they are too unique in their values and do not add any additional valuable information for our specific target. 

```python
sp = sp.drop(['Item_Identifier','Outlet_Identifier', 'Outlet_Establishment_Year'], axis=1)
sp.head()
```

```python
# X and y values
X = sp.drop('Item_Outlet_Sales', axis=1)
y = sp['Item_Outlet_Sales']
```

## Train_Test_Split

```python
# Performing a train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Making a Preprocessing Pipelines

```python
# Creating categorical and numeric selectors
# categorical selector = cat_sel
cat_sel = make_column_selector(dtype_include='object')
# numerical selector = num_sel
num_sel = make_column_selector(dtype_include='number')
```

For our provided data, we are going to use the **mean** imputer for the **numerical missing data** and the **most_frequent** imputer for the **nominal missing data**. 

As well we shall add in our StandardScaler() and OneHotEncoder() methods here.

```python
# Imputers
freq_imp = SimpleImputer(strategy='most_frequent')
mean_imp = SimpleImputer(strategy='mean')
# Scaler
scaler = StandardScaler()
# OneHotEncoder 
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)
```

```python
# Numeric pipeline
num_pipe = make_pipeline(mean_imp, scaler)
num_pipe
```

```python
# Categorical Pipeline
cat_pipe = make_pipeline(freq_imp, ohe)
cat_pipe
```

## ColumnTransformer

```python
# Tuples for Column Transformer
num_tup = (num_pipe, num_sel)
cat_tup = (cat_pipe, cat_sel)
# ColumnTransformer
preprocessor = make_column_transformer(num_tup, cat_tup, remainder='passthrough')
preprocessor
```

## Transformer Data

Now to finalize everything by processing the data and then of course, checking our work.

```python
# fit on training data
preprocessor.fit(X_train)
# transform train and test data
X_train_processed = preprocessor.transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
# Checking for missing values
print(np.isnan(X_train_processed).sum().sum())
print(np.isnan(X_test_processed).sum().sum())
```

```python
# Has the data been properly scaled and OneHotEncoded
print(X_train_processed.dtype)
print(X_test_processed.dtype)
print(X_train_processed.shape)
X_train_processed
```

## Instantiating Models

Now that we are and with cleaned and processed data, we are now ready to make some models to interpret our information.

### Function for Models

```python
# Importing metrics
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# Creating a function to represent MAE, MSE, RMSE and R2 metrics for a model
def model_metrics(pipe, X_train, y_train, X_test, y_test, model_name='Regression Model',):
  # Train
  mae = mean_absolute_error(y_train, pipe.predict(X_train))
  mse = mean_squared_error(y_train, pipe.predict(X_train))
  rmse = np.sqrt(mean_squared_error(y_train, pipe.predict(X_train)))
  r2 = r2_score(y_train, pipe.predict(X_train))
  print(f'{model_name} Train Scores')
  print(f'MAE: {mae:,.4f} \nMSE: {mse:,.4f} \nRMSE: {rmse:,.4f} \nR2: {r2:4f}\n')

  # Test
  mae = mean_absolute_error(y_test, pipe.predict(X_test))
  mse = mean_squared_error(y_test, pipe.predict(X_test))
  rmse = np.sqrt(mean_squared_error(y_test, pipe.predict(X_test)))
  r2 = r2_score(y_test, pipe.predict(X_test))

  # Display the metrics for the model
  print(f'{model_name} Test Scores')
  print(f'MAE: {mae:,.4f} \nMSE: {mse:,.4f} \nRMSE: {rmse:,.4f} \nR2: {r2:4f}\n')
```

#### Linear Regression Model

First we must take our data and impliment a pipeline for our Linear Regression Model.

```python
# Importing library
from sklearn.linear_model import LinearRegression
# Instantiating model
linreg = LinearRegression()
# Creating a model pipeline
linreg_pipe = make_pipeline(preprocessor, linreg)
# Fitting the model
linreg_pipe
```

```python
# Fitting model to our training data
linreg_pipe.fit(X_train, y_train)
```

Awesome! Now that our pipeline is all fitted, now it is time to impliment it using our model_metrics function.

```python
# Now to process the data through the Linear Regression Model
model_metrics(linreg_pipe, X_train, y_train, X_test, y_test, model_name='Linear Regression Model')
```

By what we have stated here, this model in particular is able to explain up to **50%** of both the Training and Testing data. 

With a margin of error of: 

Training Data: **$1208.86**

Testing Data: **$1168.70**

Overall Data Average error: **$1188.78**

Though that 50% model data explanation is concerning, which is why I will be doing another model as well to see if we can get a better variance of explained data and so on. 

Using Random Forests! 

#### Random Forests

```python
# Importing library
from sklearn.ensemble import RandomForestRegressor
# Instantiating model
rfreg = RandomForestRegressor()
# Creating a model pipeline
rfreg_pipe = make_pipeline(preprocessor, rfreg)
# Fitting the model
rfreg_pipe
```

```python
# Fitting model to our training data
rfreg_pipe.fit(X_train, y_train)
```

Now that our 2nd model pipeline is fitted and ready, lets crunch some numbers!

```python
# Now to process the data through the Random Forest Model
model_metrics(rfreg_pipe, X_train, y_train, X_test, y_test, model_name='Random Forest Model')
```

This model already looks far better than the first model with the testing data reporting a **53%** explanation rate, while the training data skyrocketed to a whopping **93%** explanation rate. 

Though, even with these amazing new results, it can always be better.

So lets see how we can fine tune this model

#### Model Tuning

```python
# Tuning Random Forest Model
# Tuning the max_depth for the model
est_depths = [estimator.get_depth() for estimator in rfreg.estimators_]
max(est_depths)
```

```python
# Running a range for loop to find all depths in the Random Forest Model
depths = range(1, max(est_depths))
scores = pd.DataFrame(index=depths, columns=['Train Score','Test Score'])
for depth in depths:
  model = RandomForestRegressor(max_depth=depth)
  model.fit(X_train_processed, y_train)
  scores.loc[depth, 'Train Score'] = model.score(X_train_processed, y_train)
  scores.loc[depth, 'Test Score'] = model.score(X_test_processed, y_test)
scores.head()
```

```python
# Plotting the data
plt.plot(scores['Test Score']);
```

```python
# Checking for best max_depth
sorted_scores = scores.sort_values(by='Test Score', ascending=False)
sorted_scores.head()
```

Looking over the plot graph above and then checking by sorting the data, our best depth is a depth of **5**.

Now to move on to finding our best **n_estimator**.

```python
# Tuning the estimators
n_ests = [50, 100, 150, 200, 250]
scores2 = pd.DataFrame(index=n_ests, columns=['Test Score', 'Train Score'])
for n in n_ests:
  model = RandomForestRegressor(max_depth=5, n_estimators=n)
  model.fit(X_train_processed, y_train)
  scores2.loc[n, 'Train Score'] = model.score(X_train_processed, y_train)
  scores2.loc[n, 'Test Score'] = model.score(X_test_processed, y_test)
scores2.head()
```

```python
plt.plot(scores2['Test Score'])
plt.xticks(n_ests);
```

```python
# Sorting the results to find the best score
sorted_scores2 = scores2.sort_values(by='Test Score', ascending=False)
sorted_scores2.head()
```

Now that we have found both of our optimum variables (max_depth = 5, n_estimators = 100), lets put these together and look at our finished/fine tuned model. 

```python
# Saving index value of the best test score
best_n_estimators = sorted_scores2.index[0]
# Creating a new variable to hold the fine tuning method
rfr_tuned = RandomForestRegressor(n_estimators=best_n_estimators, max_depth=5, random_state = 42)
# Fitting the new fine tunned model
rfr_tuned.fit(X_train_processed, y_train)
```

```python
# Lets plug everything together
model_metrics(rfr_tuned, X_train_processed, y_train, X_test_processed, y_test, model_name='Random Forest')
```

Now that we have the finished product before us, it is a more realistic answer. 

By what we have stated here, this model in particular is able to explain up to **60% - 61%** of both the Training and Testing data. 

With a margin of error of: 

Training Data: **$1,073.57**

Testing Data: **1,047.05**

Overall Data Average error: **$1060.31**
